```python
from transformers import BartTokenizer, BartForConditionalGeneration
import numpy as np
import pandas as pd
# Load pre-trained model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PM_Modi_speeches.csv' )
```

```python
selected_name = df.loc[1, 'text']
inputs = tokenizer.encode("summarize: " +selected_name, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = model.generate(inputs, max_length=120,min_length=80, length_penalty=2.0, num_beams=4, early_stopping=False)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
print("Word Count:", len(tokenizer.tokenize(summary)))
```
